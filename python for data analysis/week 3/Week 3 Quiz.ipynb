```python
df = pd.read_csv("datasets/cwurData.csv")

# create tier category variable
categories = ["Other Top University",
              "Third Tier Top University",
              "Second Tier Top University",
              "First Tier Top University" ]

categories = pd.CategoricalDtype(categories=categories, ordered=True)

def create_category(x):
    if (x <=100):
        return "First Tier Top University"
    elif (x > 100) & (x <= 200):
        return "Second Tier Top University"
    elif (x > 200) & (x <= 300):
        return "Third Tier Top University"
    return "Other Top University"


df['Rank_Level'] = df['world_rank'].apply(lambda x: create_category(x))


df['Rank_Level'] = df['Rank_Level'].astype(categories)
```

```python
df.pivot_table(values='score', index='country', columns='Rank_Level', aggfunc=[np.median], margins=True)
```
